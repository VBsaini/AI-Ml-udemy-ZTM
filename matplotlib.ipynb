## matplotlib example code workflow

```python
# 0) import matplot lib

%matplotlib inline
import matplotlib.pyplot as plt

# 1) prepare the data

x = [1,2,5,6]
y=[43,52,23,64]

# 2) setup data
fig,ax = plt.subplots(figsize=(10,10)) #(width, height)

# 3) plot data
ax.plot(x,y)

# 4) customize plot
ax.set(title="simple plot",
      xlabel="x-axis",
      ylabel="y-axis")

# 5) save fig
fig.savefig("./images/SP.png")
```

```python
# option 1
fig,((ax1, ax2, ax5), (ax3, ax4, ax6)) = plt.subplots(nrows=2,
                                          ncols=3,
                                          figsize=(10,5))
ax1.plot(x,x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_butter_prices.keys(),np.random.random(3))
ax4.hist(np.random.randn(1000));
```

### which method would u use(pyplot vs matplotlib oo)
* when ploting something quicly ok to use pyplot
* when ploting something advance use the OO method

```python
heartCsv = pd.read_csv("./heart-disease.csv")
heartCsv
```

```python
over50 = heartCsv[heartCsv["age"]>50]
over50.head()
```

```python
len(over50)
```

```python
# pyplot method
over50.plot(kind="scatter",
           x="age",
           y="chol",
           c="target")
```

```python
# OO method
fix, ax = plt.subplots(figsize=(10,6))
over50.plot(kind="scatter",x="age",y="chol",c="target",ax=ax);
# ax.set_xlim([45,100])
```

```python
## OO method
fig, ax = plt.subplots(figsize=(10,6))
scatter = ax.scatter(x=over50["age"], y=over50["chol"], c=over50["target"]);
ax.set_xlim([45,80])
ax.set(title="Heart Diesease and Cholestrol levers", xlabel="Age", ylabel="Cholestrol")
ax.legend(*scatter.legend_elements(), title="target")
ax.axhline(over50["chol"].mean(), linestyle="--")
```

# suplot of chol, age, thalach

```python
plt.style.use("classic")
fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10,10))
scatterChol = ax0.scatter(x=over50["age"], y=over50["chol"], c=over50["target"]);
ax0.set_xlim([45,80])
ax0.set(title="Heart Diesease and Cholestrol levers", xlabel="Age", ylabel="Cholestrol")
ax0.legend(*scatterChol.legend_elements(), title="target")
ax0.axhline(over50["chol"].mean(), linestyle="--")

scatterThalach = ax1.scatter(x=over50["age"], y=over50["thalach"], c=over50["target"]);
ax1.set_xlim([45,80])
ax1.set(title="Heart Diesease and Cholestrol levers", xlabel="Age", ylabel="Max hearteat")
ax1.legend(*scatterThalach.legend_elements(), title="target")
ax1.axhline(over50["thalach"].mean(), linestyle="--")
fig.suptitle("Heart Disease Analysis", fontsize="16", fontweight="bold");
```

```python
x=np.random.randn(10,4)
df=pd.DataFrame(x,columns=['a','b','c','d'])
ax=df.plot(kind="bar")
ax.set(title="random number bar graph from dataframe",
       xlabel="row number",
       ylabel="random number")
ax.legend().set_visible(True)
```

```python
plt.style.use("seaborn-v0_8-whitegrid")
fig, ax = plt.subplots(figsize=(10,6))
scatter = ax.scatter(x=over50["age"], 
                     y=over50["chol"], 
                     c=over50["target"], 
                     cmap="winter", # this changes the color scheme 
                     #s=[80]
                    );
ax.set(title="Heart Diesease and Cholestrol levers", xlabel="Age", ylabel="Cholestrol")
ax.legend(*scatter.legend_elements(), title="target")
ax.axhline(over50["chol"].mean(), linestyle="--");
```

```python
fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10,10))
scatterChol = ax0.scatter(x=over50["age"], y=over50["chol"], c=over50["target"], cmap="winter", s=[70]);
ax0.set_xlim([50,80])
# ax0.set_ylim([45,80])
ax0.set(title="Heart Diesease and Cholestrol levers", xlabel="Age", ylabel="Cholestrol")
ax0.legend(*scatterChol.legend_elements(), title="target")
ax0.axhline(over50["chol"].mean(), linestyle="--", c=(1,0,0), linewidth = 2)

scatterThalach = ax1.scatter(x=over50["age"], y=over50["thalach"], c=over50["target"], cmap="winter", s=[70]);
ax1.set_xlim([50,80])
ax1.set_ylim([60,200])
ax1.set(title="Heart Diesease and Cholestrol levers", xlabel="Age", ylabel="Max hearteat")
ax1.legend(*scatterThalach.legend_elements(), title="target")
ax1.axhline(over50["thalach"].mean(), linestyle="--", c=(1,0,0), linewidth = 2)
fig.suptitle("Heart Disease Analysis", fontsize="16", fontweight="bold");
```
